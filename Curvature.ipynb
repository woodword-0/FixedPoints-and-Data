```python
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
% matplotlib notebook

# Example function
def f(x, y, t):
    return np.sin(np.sqrt(x ** 2 + y ** 2 + t))

####################################################################################
####################################################################################
# B.S. Data
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Create the figure and animation
fig = plt.figure()
ax = plt.axes(projection='3d')
line = ax.plot_surface(X, Y, f(X, Y, 0), cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

def update(t):
    line._offsets3d = (X.flatten(), Y.flatten(), f(X, Y, t).flatten())

anim = FuncAnimation(fig, update, frames=np.linspace(0, 10, 100), interval=100)
plt.show()
# Display the animation in the notebook
# HTML(anim.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection="3d")

# Define the data
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t

# Create a line object
line, = ax.plot(x, y, z)

# Define the update function
def update(num):
    # Rotate the data by 0.1 radians
    line.set_data(np.cos(t + 0.1*num), np.sin(t + 0.1*num))
    line.set_3d_properties(t + 0.1*num)
    return line,

# Create the animation
anim = FuncAnimation(fig, update, frames=100, interval=50)

# Display the animation
plt.show()
```
